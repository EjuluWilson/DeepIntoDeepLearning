```python
import numpy as np
```

```python
# Inputs
X = np.array([[1],  # x1
              [2]]) # x2

# Weights for the hidden layer
W1 = np.array([[0.2, 0.4],  # Weights for h1 and h2 from x1
               [0.5, 0.3]]) # Weights for h1 and h2 from x2

# Bias for the hidden layer
b1 = np.array([[0.1],  # Bias for h1
               [-0.2]]) # Bias for h2

# Weights for the output layer
W2 = np.array([[0.7],  # Weight for output neuron from h1
               [0.6]]) # Weight for output neuron from h2

# Bias for the output layer
b2 = np.array([[0.3]]) # Bias for the output neuron
```

```python
print(X)
```

```python
# define activation functions 
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Compute z for the hidden layer
z1 = W1.T @ X + b1  # W1.T for the transposed weight matrix

# Apply ReLU activation
a1 = relu(z1)

# Print results for verification
print("z1 (hidden layer pre-activation):\n", z1)
print("a1 (hidden layer activation):\n", a1)
```

```python
#final]
# Compute z for the output layer
z2 = W2.T @ a1 + b2  # W2.T for the transposed weight matrix

# Apply Sigmoid activation
y_hat = sigmoid(z2)

# Print results for verification
print("z2 (output layer pre-activation):", z2)
print("y_hat (final prediction):", y_hat)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (positive values only, as log is undefined for x <= 0)
x = np.linspace(0.1, 10, 500)  # Start from a small positive value

# Calculate logarithmic values for different bases
y_base_e = np.log(1-x)           # Natural logarithm, base e
y_base_10 = np.log10(x)        # Logarithm, base 10
y_base_2 = -np.log(x) / np.log(2)#np.log2(x)          # Logarithm, base 2

# Plot the logarithmic functions
plt.figure(figsize=(10, 6))
plt.plot(x, y_base_e, label='log_e(x) (Natural Log)', linewidth=2)
# plt.plot(x, y_base_10, label='log_10(x)', linewidth=2)
# plt.plot(x, y_base_2, label='log_2(x)', linewidth=2)

# Add labels and title
plt.title('Logarithmic Functions', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('log_b(x)', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add y=0 axis
plt.axvline(1, color='black', linewidth=0.8, linestyle='--')  # Add x=1 axis
plt.legend(fontsize=12)
plt.grid(alpha=0.5)

# Show the plot
plt.show()
```
